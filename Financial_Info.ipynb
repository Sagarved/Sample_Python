## __Information about the Company from Yfinance response__
1. Dividend Rate and Dividend Yield

    1. Dividend Rate

        The dividend rate is the amount of money a company pays in dividends per share per year.

        In your example:
        dividendRate: 1.0 means the company pays $1.00 in dividends annually for each share you own.

    2. Dividend Yield

        The dividend yield is the dividend as a percentage of the current stock price.

        It helps investors understand the return they are getting from dividends relative to the stock price.

        Dividend Yield = (Dividend in Dollar / Stock price in Dollar) * 100

        In your example:
        dividendYield: 0.51 means the yield is 0.51%.
2. Payout Ratio 

    The payout ratio tells you what portion of a company’s net income is paid out to shareholders as dividends.
    
    Payout Ratio = (Dividends per Share / Earnings per Share​) * 100

    __A low payout ratio (like 10-40%) can mean:__

        The company is reinvesting most of its profits for growth.

        The dividend is likely sustainable even if profits fluctuate.

    __A high payout ratio (e.g., 80–100%) might be riskier:__

        Less money is being kept for growth or emergencies.

        Dividends might be cut if earnings fall.
3. Beta: Beta shows how much a stock moves relative to a benchmark index (S&P 500).

    __1.0__	Stock moves in line with the market. If the market goes up 1%, the stock likely goes up 1%.

    __> 1.0__	Stock is more volatile than the market. (e.g., Beta = 1.5 → stock tends to move 1.5% for every 1% market move).

    __< 1.0__	Stock is less volatile than the market. (e.g., Beta = 0.5 → stock moves only 0.5% when market moves 1%).

    __0__	No correlation to the market. Often cash or very stable assets.

    __Negative__ Beta	Stock moves opposite to the market. Rare, but some assets (like gold) may behave like this.

4. __Market-cap vs Enterprise Value__

   * Market cap is multiplying outstanding shares with current price to gauge mid-cap, large-cap size of the company.

   * Enterprise Value (EV) is more accurate way to measure company's valuation, such as if one wants to purchase company how much it is costing.

        EV = Market Cap __+__ Total Debt __−__ Cash and Cash Equivalents

5. Profit Margin: Net Income / Revenue: It shows how lucrative is business that how much of every dollar of revenue, the company keep it as profit after all the expenses.

6. Shares Outstanding: This is the total number of shares a company has __issued and are held by all shareholders__ (including insiders, institutions, and the public).

    It includes both freely traded and restricted shares.   

7. __FloatShares__: These are the shares that are available to trade on the open market.

    🔹 Lower float = More volatility
    🔹 Higher float = More stability

8. __sharesShort__: This is the number of shares that investors have borrowed and sold, betting the stock price will go down.

    Short Floats = (shareShort/FloatShares) * 100 

    Short Floats = (1000/100000) * 100 = 1, means market is not bearish on this stock

9. __Book Value__ = (Total Assets - Total Liabilities​)/ Shares Outstanding 

    It tells you how much the company would be worth if it liquidated today — a conservative view of value.

10. __priceToBook__= Current Share Price / Book Value per Share

    * __P/B Value	Interpretation__

    1. < 1	Stock is trading below book value → may be undervalued

    2. 1 - 3	Reasonable for many sectors

    3. Greater than 3	Investors are paying a premium over book value

    * P/B is more meaningful for asset-heavy industries (like banks, manufacturing).

    * Less useful for tech or service companies where most value is intangible (e.g., brand, IP, user base).

```python
# Extracting key metrics for Warren Buffett-style analysis
net_income = financials.loc['Net Income', :].astype(float)
operating_income = financials.loc['Operating Income', :].astype(float)
ebitda = financials.loc['EBITDA', :].astype(float)
eps = financials.loc['Diluted EPS', :].astype(float)
roe = info['returnOnEquity']
roa = info['returnOnAssets']
profit_margin = info['profitMargins']
total_debt = balance_sheet.loc['Total Debt', :].astype(float)
net_debt = balance_sheet.loc['Net Debt', :].astype(float)
debt_to_equity = info['debtToEquity']
free_cash_flow = cashflow.loc['Free Cash Flow', :].astype(float)
current_ratio = info['currentRatio']
quick_ratio = info['quickRatio']
pe_ratio = info['trailingPE']
pb_ratio = info['priceToBook']
ps_ratio = info['priceToSalesTrailing12Months']
dividend_yield = info['dividendYield']
dividend_payout_ratio = info['payoutRatio']
revenue_growth = info['revenueGrowth']
earnings_growth = info['earningsGrowth']
operating_cash_flow = cashflow.loc['Operating Cash Flow', :].astype(float)
cash_and_cash_equivalents = balance_sheet.loc['Cash And Cash Equivalents', :].astype(float)
shares_outstanding = balance_sheet.loc['Ordinary Shares Number', :].astype(float)
share_buybacks = cashflow.loc['Repurchase Of Capital Stock', :].astype(float)

# Printing the extracted metrics
print("Net Income:", net_income)
print("Operating Income:", operating_income)
print("EBITDA:", ebitda)
print("EPS:", eps)
print("ROE:", roe)
print("ROA:", roa)
print("Profit Margin:", profit_margin)
print("Total Debt:", total_debt)
print("Net Debt:", net_debt)
print("Debt to Equity Ratio:", debt_to_equity)
print("Free Cash Flow:", free_cash_flow)
print("Current Ratio:", current_ratio)
print("Quick Ratio:", quick_ratio)
print("P/E Ratio:", pe_ratio)
print("P/B Ratio:", pb_ratio)
print("P/S Ratio:", ps_ratio)
print("Dividend Yield:", dividend_yield)
print("Dividend Payout Ratio:", dividend_payout_ratio)
print("Revenue Growth:", revenue_growth)
print("Earnings Growth:", earnings_growth)
print("Operating Cash Flow:", operating_cash_flow)
print("Cash and Cash Equivalents:", cash_and_cash_equivalents)
print("Shares Outstanding:", shares_outstanding)
print("Share Buybacks:", share_buybacks)
```

```python
# Calculate key financial ratios
ratios = {
    'ROE': roe,
    'ROA': roa,
    'Profit Margin': profit_margin,
    'Debt to Equity Ratio': debt_to_equity,
    'Current Ratio': current_ratio,
    'Quick Ratio': quick_ratio,
    'P/E Ratio': pe_ratio,
    'P/B Ratio': pb_ratio,
    'P/S Ratio': ps_ratio,
    'Dividend Yield': dividend_yield,
    'Dividend Payout Ratio': dividend_payout_ratio,
    'Revenue Growth': revenue_growth,
    'Earnings Growth': earnings_growth
}

# Print the ratios and provide comments in Warren Buffett style
for ratio, value in ratios.items():
    print(f"{ratio}: {value}")
    if ratio == 'ROE':
        print("Comment: A high ROE indicates that the company is efficient in generating profits from shareholders' equity. Warren Buffett prefers companies with an ROE above 15%.")
    elif ratio == 'ROA':
        print("Comment: A high ROA indicates that the company is efficient in generating profits from its assets. Warren Buffett looks for companies with a strong ROA.")
    elif ratio == 'Profit Margin':
        print("Comment: A high profit margin indicates that the company is good at converting revenue into actual profit. Warren Buffett prefers companies with a consistent and high profit margin.")
    elif ratio == 'Debt to Equity Ratio':
        print("Comment: A low debt to equity ratio indicates that the company is not heavily reliant on debt to finance its growth. Warren Buffett prefers companies with a debt to equity ratio below 0.5.")
    elif ratio == 'Current Ratio':
        print("Comment: A current ratio above 1 indicates that the company has enough short-term assets to cover its short-term liabilities. Warren Buffett prefers companies with a strong current ratio.")
    elif ratio == 'Quick Ratio':
        print("Comment: A quick ratio above 1 indicates that the company can meet its short-term obligations without relying on inventory sales. Warren Buffett looks for companies with a strong quick ratio.")
    elif ratio == 'P/E Ratio':
        print("Comment: A lower P/E ratio may indicate that the stock is undervalued. Warren Buffett looks for companies with a reasonable P/E ratio compared to their growth prospects.")
    elif ratio == 'P/B Ratio':
        print("Comment: A lower P/B ratio may indicate that the stock is undervalued. Warren Buffett prefers companies with a P/B ratio below 3.")
    elif ratio == 'P/S Ratio':
        print("Comment: A lower P/S ratio may indicate that the stock is undervalued. Warren Buffett looks for companies with a reasonable P/S ratio compared to their growth prospects.")
    elif ratio == 'Dividend Yield':
        print("Comment: A higher dividend yield indicates that the company returns a significant portion of its earnings to shareholders. Warren Buffett prefers companies with a stable and growing dividend yield.")
    elif ratio == 'Dividend Payout Ratio':
        print("Comment: A lower dividend payout ratio indicates that the company retains more earnings for growth. Warren Buffett prefers companies with a sustainable dividend payout ratio.")
    elif ratio == 'Revenue Growth':
        print("Comment: Consistent revenue growth indicates that the company is expanding its business. Warren Buffett looks for companies with a strong and consistent revenue growth.")
    elif ratio == 'Earnings Growth':
        print("Comment: Consistent earnings growth indicates that the company is increasing its profitability. Warren Buffett prefers companies with a strong and consistent earnings growth.")
    print()
```

```python
# Calculate Enterprise Value
market_cap = info['marketCap']
total_debt = balance_sheet.loc['Total Debt', :].astype(float).iloc[0]
cash_and_cash_equivalents = balance_sheet.loc['Cash And Cash Equivalents', :].astype(float).iloc[0]

enterprise_value = market_cap + total_debt - cash_and_cash_equivalents

# Print the Enterprise Value and provide a comment
print(f"Enterprise Value: {enterprise_value}")
print("Comment: Warren Buffett prefers companies with a strong enterprise value, as it provides a more comprehensive measure of a company's total value. A lower enterprise value relative to earnings indicates a potentially undervalued company.")
```
